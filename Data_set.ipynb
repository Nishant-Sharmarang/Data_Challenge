```python
from sklearn.model_selection import train_test_split
X = modelling_data_without_profit
y = modelling_data_with_profit['Profit']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
reg = LinearRegression()
reg.fit(X_train, y_train)
y_predicted = reg.predict(X_test)
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_predicted))
print('RÂ²: %.2f' % r2_score(y_true=y_test, y_pred=y_predicted))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
regressor = RandomForestRegressor(n_estimators=500, random_state=0)  
regressor.fit(X_train, y_train)  
y_pred = regressor.predict(X_test)
```

```python
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
print('RÂ²: %.2f' % r2_score(y_test, y_pred))
```

```python
print ("Features sorted by their score:")
print (sorted(zip(map(lambda x: round(x, 4), regressor.feature_importances_),modelling_data_without_profit ), 
             reverse=True))
```
